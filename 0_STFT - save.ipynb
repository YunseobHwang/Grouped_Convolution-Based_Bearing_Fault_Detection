```python
def printProgress(iteration, total, prefix = '', suffix = '', decimals = 1, barLength = 100):
    formatStr = "{0:." + str(decimals) + "f}"
    percent = formatStr.format(100 * (iteration / float(total)))
    filledLength = int(round(barLength * iteration / float(total)))
    bar = '#' * filledLength + '-' * (barLength - filledLength)
    sys.stdout.write('\r{} |{} | {}{} {}'.format(prefix, bar, percent, '%', suffix)),
    if iteration == total:
        sys.stdout.write('\n')
    sys.stdout.flush()

def save_STFT(data, file_name):
    """
    example
    data shape: (4,)
    data 1st: (121846,)
    data 2nd: (122136,)
    data 3rd: (121991,)
    data 4th: (122136,) 
    """
    print(file_name)
    image = np.zeros([1000, 100, 100])
    for j in range(4):
        for i in range(250):
            image[j
                  *250 + i, :, :] = np.abs(librosa.stft(data[j], n_fft = 199, hop_length = None, win_length = 50))[:, 40*i:40*i+100]
            printProgress(i + 1, 250, j + 1)

    image = np.expand_dims(image, axis = 3)
    np.save('/mnt/disk1/yunseob/courses/19-2_computer vision/data/STFT/' + file_name, image)
    
    del image
```

```python
save_STFT(normal, 'normal')
save_STFT(ball_7, 'ball_7')
save_STFT(ball_14, 'ball_14')
save_STFT(ball_21, 'ball_21')
```
